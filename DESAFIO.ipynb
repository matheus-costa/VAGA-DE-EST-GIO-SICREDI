1. O objetivo deste desafio é criar um fluxo de trabalho que insira
1.1 dados de uma planilha nos campos do formulário na tela.

   32. Cuidado! Os campos mudarão de posição na tela após cada envio a
2.1  longo de 10 rodadas, portanto o fluxo de trabalho deve identifica
    corretamente onde cada registro da planilha deve ser digitado a cada vez.

```python

from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium import webdriver
import pyautogui
import time

pyautogui.pause = 0.5

navegador = webdriver.Chrome()
navegador.get("https://www.rpachallenge.com/")


navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[7]/rpa1-field/div/input').send_keys("Qualquer nome")
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[3]/rpa1-field/div/input').send_keys("Qualquer empresa")
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[2]/rpa1-field/div/input').send_keys(" Qualquer endereco")
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[5]/rpa1-field/div/input').send_keys("Qualquer da empresa")
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[4]/rpa1-field/div/input').send_keys("Qualquer do telefone")
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[6]/rpa1-field/div/input').send_keys("Qualquer sobrenome")
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[1]/rpa1-field/div/input').send_keys("Qualquer@gmail.com")
time.sleep(2)


navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]/form/input').click()

time.sleep(3)

import pandas as pd

tabela = pd.read_excel("challenge.xlsx")
display(tabela)

for i, primeiroNome in enumerate(tabela ["First Name"]):
       segundoNome = tabela.loc[i, "Last Name"]
       nomeEmpresa = tabela.loc[i, "Company Name"]
       funcaoEmpresa = tabela.loc[i, "Role In Company"]
       endereco = tabela.loc[i, "Adress"]
       email = tabela.loc[i, "Email"]
       telefone = tabela.loc[i, "Phone Number"]

    
time.sleep(1)
   
navegador.get("https://www.rpachallenge.com/")


navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[7]/rpa1-field/div/input').send_keys(primeiroNome)                    
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[3]/rpa1-field/div/input').send_keys(segundoNome)
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[2]/rpa1-field/div/input').send_keys(nomeEmpresa)
                    
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[5]/rpa1-field/div/input').send_keys(funcaoEmpresa)
                      
time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[4]/rpa1-field/div/input').send_keys(endereco)

time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[6]/rpa1-field/div/input').send_keys(email)

time.sleep(1)
navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]'
                       '/form/div/div[1]/rpa1-field/div/input').send_keys(telefone)

time.sleep(1)

navegador.find_element("xpath",'/html/body/app-root/div[2]/app-rpa1/div/div[2]/form/input').click()
```
